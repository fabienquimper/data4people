```python
import polars as pl
import numpy as np

# PERMET D'EXTRAIRE DES LIEUX AUTOUR D'UN POINT GPS

# Paramètres
fichier = "../data/raw/osm/full.csv/full.csv"
# Ex: Mairie de Quimper
lat0 = 47.996197
lon0 = -4.102031
rayon_m = 50 # Distance en mètre
cp_filtre = "29000"

str_schema = {"id": pl.String, "numero": pl.String, "voie": pl.String, "cp": pl.String, "commune": pl.String, "source": pl.String, "lat": pl.Float32, "lon": pl.Float32}

osm_df = pl.read_csv(
    "../data/raw/osm/full.csv/full.csv",
    infer_schema_length=0,
    try_parse_dates=False,
    schema_overrides=str_schema
)

# ------------------------------
# Filtrer par code postal (évite une surcharge)
# ------------------------------
df_cp = osm_df.filter(pl.col("cp") == cp_filtre)

# ------------------------------
# Calcul distance approximative en km
# ------------------------------
df_distance = df_cp.with_columns([
    (
        ((pl.col("lat").cast(pl.Float64) - lat0) * 111)**2 +
        ((pl.col("lon").cast(pl.Float64) - lon0) * 111 * np.cos(np.radians(lat0)))**2
    ).sqrt().alias("distance_km")
])

# ------------------------------
# Filtrer par rayon
# ------------------------------
df_rayon = df_distance.filter(pl.col("distance_km") <= (rayon_m / 1000))

# ------------------------------
# Afficher résultats
# ------------------------------
res = df_rayon.head(10)
print(res)
# print(res.to_dicts())
```
