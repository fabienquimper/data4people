```python
import polars as pl
import plotly.express as px

df = pl.read_csv("../data/raw/234400034_070-003_offre-touristique-hebergements-collectifs-rpdl@paysdelaloire.csv", separator=";")

df = df.with_columns([
    pl.col("Latitude").cast(pl.Float64).alias("lat"),
    pl.col("Longitude").cast(pl.Float64).alias("lon")
]).drop_nulls(["Latitude", "Longitude"])

```

```python
pdf = df.to_pandas()

print(pdf.head())

fig = px.scatter_mapbox(
    pdf,
    lat="lat",
    lon="lon",
    color="Type de l'offre",
    hover_name="Nom de l'offre touristique",
    zoom=6,
    height=700
)









```

```python
fig.update_layout(mapbox_style="open-street-map")
fig.show()
```
