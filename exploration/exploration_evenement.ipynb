```python
import pandas as pd

# 🔹 Charger le CSV final une seule fois
CSV_PATH = "../llm_server/table_correspondance_categories2.csv"
df = pd.read_csv(CSV_PATH, sep=";")

def get_neutral_category(dataset_name: str, category_name: str):
    """
    Retourne la catégorie neutre et l'alias pour une catégorie d'un dataset donné.

    Args:
        dataset_name (str): 'pdl' ou 'bzh' (ou autre nom figurant dans la colonne 'dataset')
        category_name (str): nom exact de la catégorie du dataset

    Returns:
        dict: {'categorie_neutre': ..., 'alias': ...} ou None si non trouvé
    """
    row = df[(df['dataset'] == dataset_name) & (df['categorie_dataset'] == category_name)]

    if row.empty:
        return None

    # On prend la première correspondance trouvée
    categorie_neutre = row.iloc[0]['categorie_neutre']
    alias = row.iloc[0]['alias']

    return {
        'categorie_neutre': categorie_neutre,
        'alias': alias
    }

# 🔹 Exemple d'utilisation
res = get_neutral_category("pdl", "Cirque")
print(res)  # {'categorie_neutre': 'Cirque', 'alias': 'cirque'}

res2 = get_neutral_category("bzh", "Culturelle#Musique")
print(res2)  # {'categorie_neutre': 'Musique', 'alias': 'musique'}
```

```python
# -----------------------------
# 🔹 Charger la table de correspondance CSV
# -----------------------------
df_map = pd.read_csv("../llm_server/table_correspondance_categories2.csv", sep=";")

def get_neutral_category(dataset_name: str, category_name: str):
    """Récupère la catégorie neutre et l'alias depuis le CSV de mapping"""
    row = df_map[(df_map['dataset'] == dataset_name) & (df_map['categorie_dataset'] == category_name)]
    if row.empty:
        return category_name  # si pas trouvé, renvoyer la catégorie originale
    return row.iloc[0]['categorie_neutre']

# -----------------------------
# 🔹 Charger et normaliser les datasets
# -----------------------------
# Pays de la Loire
df_pdl = pl.read_csv(
    "../data/raw/234400034_070-002_offre-touristique-fetes_et_manifestations-rpdl@paysdelaloire.csv",
    separator=";"
).with_columns([
    pl.col("Latitude").cast(pl.Float64).alias("lat"),
    pl.col("Longitude").cast(pl.Float64).alias("lon"),
    pl.col("Catégorie de la fête ou manifestation").alias("category"),
    pl.col("Nom de la fête ou manifestation").alias("name")
]).drop_nulls(["lat", "lon"]).select(["lat", "lon", "category", "name"])

# Bretagne
df_bretagne = pl.read_csv(
    "../data/raw/bretagne-fetes-et-manifestations.csv",
    separator=";",
    schema_overrides={
        "DetailIDENTADRESSECP": pl.Utf8,
        "DetailIDENTFMATYPE": pl.Utf8,
        "SyndicObjectName": pl.Utf8,
        "GmapLatitude": pl.Float64,
        "GmapLongitude": pl.Float64
    },
    ignore_errors=True
).with_columns([
    pl.col("GmapLatitude").alias("lat"),
    pl.col("GmapLongitude").alias("lon"),
    pl.col("DetailIDENTFMATYPE").alias("category"),
    pl.col("SyndicObjectName").alias("name")
]).drop_nulls(["lat", "lon"]).select(["lat", "lon", "category", "name"])

# -----------------------------
# 🔹 Ajouter la catégorie neutre
# -----------------------------
def add_neutral_category(pl_df, dataset_code):
    pdf = pl_df.to_pandas()
    pdf['neutral_category'] = pdf['category'].apply(lambda c: get_neutral_category(dataset_code, c))
    return pdf

pdf_pdl = add_neutral_category(df_pdl, 'pdl')
pdf_bzh = add_neutral_category(df_bretagne, 'bzh')

# -----------------------------
# 🔹 Fusionner
# -----------------------------
pdf_all = pd.concat([pdf_pdl, pdf_bzh], ignore_index=True)

# -----------------------------
# 🔹 Carte avec catégorie neutre
# -----------------------------
fig = px.scatter_mapbox(
    pdf_all,
    lat="lat",
    lon="lon",
    color="neutral_category",  # <-- utiliser la catégorie neutre
    hover_name="name",
    zoom=6,
    height=800
)

fig.update_layout(mapbox_style="open-street-map")
fig.show()
```

```python
# Après avoir créé la figure `fig`
fig.write_html("../export_html/carte_interactive_evenements.html", include_plotlyjs="cdn")

fig.update_layout(
    width=2000,   # largeur en pixels
    height=1200   # hauteur en pixels
)

fig.write_image("../export_html/carte_interactive_evenements.png", scale=1)
```
