```python
import polars as pl

VILLE_CODE_INSEE = "29232"

file_comer = "../data/raw/cog_ensemble_2025_csv/v_commune_comer_2025.csv"
file = "../data/raw/cog_ensemble_2025_csv/v_commune_2025.csv"
print("Source: https://www.insee.fr/fr/information/2560452")

df = pl.read_csv(file,
        schema_overrides={
        "COM": pl.String,
        "DEP": pl.String,
        "REG": pl.String,
        "COM_COMER": pl.String,   # si présent dans ce fichier
        }
    )
df35 = df.filter(pl.col("COM") == VILLE_CODE_INSEE)
# print(df.head(3).to_pandas())

print("Nom Ville via CODE INSEE ", VILLE_CODE_INSEE)
nom_ville = df35[0]["LIBELLE"]
nom_ville
```

```python
# Détail ville INSEE

dfComer = pl.read_csv(file_comer)
dfComer

dfComerElt = df.filter(pl.col("LIBELLE") == nom_ville)
print(dfComerElt.to_dicts())
dfComerElt

# Pas de code postal ici, justa la "ville associée"
```
